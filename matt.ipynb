```python
#For each province, we want to generate a TCN prediction model for the MW demand for each generation type
# We will Eventually try to predict all the way to 2030 for each province
#Lets try to predict the MW demand for the last year of available data for each province
# We will use all of the data up to that point to train the model and validation data to test the model
# lets start with Alberta

# Alberta
#use all fo the data up to 2021 to train the model

df_ab = df_ab[df_ab['date'] < '2022-01-01']
df_ab = df_ab.groupby(['date', 'generation_type'], as_index=False)['megawatt_hours'].sum()

#split the data into training and validation data
train = df_ab[df_ab['date'] < '2021-01-01']
valid = df_ab[df_ab['date'] >= '2021-01-01']

# get the test data
test = df_ab[df_ab['date'] >= '2022-01-01']

#convert the date column to datetime
train['date'] = pd.to_datetime(train['date'])
valid['date'] = pd.to_datetime(valid['date'])
test['date'] = pd.to_datetime(test['date'])

#set date as the index
train.index = train['date']
valid.index = valid['date']
test.index = test['date']

#drop the date column
train.drop('date', axis=1, inplace=True)
valid.drop('date', axis=1, inplace=True)
test.drop('date', axis=1, inplace=True)



```
